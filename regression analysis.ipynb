# Analysis of weather data set (Szeged)

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as mp

df.columns = ["date","summary","precip","temp","apparent_temp","humidity","wind_speed","wind_bearing","visibility","loud","pressure","daily_summary"]
df.head()
```

```
# between ap_temp and temperature
df.dropna(inplace = True)
ap_temp = np.array( df['apparent_temp'])
temp = np.array(df['temp'])

mp.scatter(ap_temp,temp,s=10)
mp.xlabel('ap_temp')
mp.ylabel('temperature')
mp.show()
```

```
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

regression_model = LinearRegression()
# Fit the data(train the model)
regression_model.fit(ap_temp.reshape(-1, 1), temp.reshape(-1, 1))
# Predict
y_predicted = regression_model.predict(ap_temp.reshape(-1, 1))

mp.scatter(ap_temp,temp,s=10)
mp.xlabel('ap_temp')
mp.ylabel('temperature')
mp.plot(ap_temp, y_predicted, color='r')
mp.show()
```
